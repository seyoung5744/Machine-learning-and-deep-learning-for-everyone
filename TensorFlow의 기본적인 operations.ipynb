# Computational Graph





```python
node1 = tf.constant(3.0, tf.float32)
node2 = tf.constant(4.0) # also tf.float32 implicitly
node3 = tf.add(node1, node2)
```

```python
print("node1:", node1, "node2:",node2)
print("node3:", node3)
```

```python
sess = tf.compat.v1.Session()
print("sess.run(node1, node2):", sess.run([node1, node2]))
print("sess.run(node3):", sess.run(node3))
```



-  graph 빌드
-  sess.run(op)을 통해서 graph 실행
-  값 return or graph 업데이트

### (1) Build graph (tensors) using TensorFlow operaions

```python
node1 = tf.constant(3.0, tf.float32)
node2 = tf.constant(4.0) # also tf.float32 implicitly
node3 = tf.add(node1, node2)
```

### (2) feed data and run graph (operation) sess.run(op)
### (3) update variables in the graph (and return values)

```python
sess = tf.compat.v1.Session()
print("sess.run(node1, node2):", sess.run([node1, node2]))
print("sess.run(node3):", sess.run(node3))
```

# Placeholder

위의 과정은 그래프에 상수를 지정하고 run을 하였다. 그렇다면 그래프는 미리 만들어두고 실행시키는 단계에서 값들을 넣고 싶다면??<br>
이때 node를 만들지만 node를 placeholder라는 노드로 만든다.

```python
a = tf.compat.v1.placeholder(tf.float32)
b = tf.compat.v1.placeholder(tf.float32)
adder_node = a + b # + provides a shortcut for tf.add(a, b)

print(sess.run(adder_node, feed_dict={a: 3, b: 4.5}))
print(sess.run(adder_node, feed_dict={a: [1,3], b: [2,4]}))
```
